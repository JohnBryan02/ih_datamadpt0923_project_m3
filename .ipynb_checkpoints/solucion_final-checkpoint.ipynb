#FEATURES

```python
diamonds = pd.read_csv('./data/diamondsql.csv')
#diamonds['volume'] = (4/3) * np.pi * diamonds['x'] * diamonds['y'] * diamonds['z']
diamonds['length_width_ratio'] = diamonds['x'] / diamonds['y']
diamonds['width_depth_ratio'] = diamonds['y'] / diamonds['z']
diamonds['symmetry_index'] = (diamonds['x'] * diamonds['y'] * diamonds['z']) / (diamonds['x'] + diamonds['y'] + diamonds['z'])
diamonds['brilliance_index'] = diamonds['carat'] / (diamonds['x'] * diamonds['y'] * diamonds['z'])
cut_score = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}
diamonds['cut_score'] = diamonds['cut'].map(cut_score)
diamonds['proportion_index'] = (diamonds['x'] / diamonds['y']) * (diamonds['x'] / diamonds['z'])
diamonds['purity_index'] = (diamonds['clarity'].astype('category').cat.codes + diamonds['color'].astype('category').cat.codes) / 2
diamonds['is_from_dubai'] = np.where(diamonds['city'] == 'Dubai', 1, 0)
diamonds['overall_quality'] = diamonds[['symmetry_index', 'brilliance_index', 'cut_score', 'proportion_index', 'purity_index']].mean(axis=1)
#diamonds
```

LabelEncoder,OneHotEncoding, getdummies

```python
#Tras varios test he decidido quedarme con LabelEncoder
label_encoder = LabelEncoder()
```

```python
diamonds.info() #Revisamos que columnas no son númericas
```

```python
#Concluimos que son estas ya que salen con type(object)
columns_to_encode = ['cut','clarity','color']
```

```python
diamonds_encoded = diamonds.copy()  # Hacemos una copia de diamonds original para no modificarlo directamente
for column in columns_to_encode: #Con este for el label_encoder se aplicara en las columnas que hemos puesto en columns_to_encode
    diamonds_encoded[column] = label_encoder.fit_transform(diamonds[column]) 
```

Data vs Target

```python
X = diamonds_encoded.drop('price', axis=1)
y = diamonds_encoded['price']
```

```python
X.head()
```

```python
%%time
model_xgb = XGBRegressor(missing=np.inf) #usamos XGBRegressor ya que es la que mejor ha estado funcionando.
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [1000, 1500],
    'subsample': [0.7, 0.8, 0.9],
    'max_depth': [7],
    'colsample_bytree': [0.7, 0.8],
    'random_state' : [42]
}     #Denifimos los hiperparametros y vamos jugando con ellos hasta conseguir el mejor SCORE!
grid_search = GridSearchCV(estimator=model_xgb, param_grid=param_grid, cv=5, n_jobs=-1, verbose=3, scoring='neg_root_mean_squared_error')
grid_search.fit(X, y) #fitting!
best_params = grid_search.best_params_
best_score = grid_search.best_score_
best_model = grid_search.best_estimator_
#Imprimimos para poder ir viendo cuales son los mejores hiperparametros y que podemos ir descartando
print("Best parameters found: ", best_params)
print("Best score found: ", best_score)
```
